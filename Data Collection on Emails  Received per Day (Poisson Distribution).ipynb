# **Data Collection on Emails per day**

## Step 1: Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

## Step 2: Define the Dataset

```python
emails_data = {
    "Person A": {"Monday": 9, "Tuesday": 11, "Wednesday": 10, "Thursday": 13, "Friday": 8, "Saturday": 12, "Sunday": 10},
    "Person B": {"Monday": 8, "Tuesday": 10, "Wednesday": 9, "Thursday": 11, "Friday": 7, "Saturday": 8, "Sunday": 9},
    "Person C": {"Monday": 10, "Tuesday": 12, "Wednesday": 11, "Thursday": 14, "Friday": 9, "Saturday": 13, "Sunday": 12},
}
```

## Step 3: Display the Dataset

```python
print("\nDataset: Emails Received per Person per Day\n")
print("| Person   | Monday | Tuesday | Wednesday | Thursday | Friday | Saturday | Sunday |")
print("|----------|--------|---------|-----------|----------|--------|----------|--------|")
for person, days in emails_data.items():
    print(f"| {person:<8} | {days['Monday']:<6} | {days['Tuesday']:<7} | {days['Wednesday']:<9} | {days['Thursday']:<8} | {days['Friday']:<6} | {days['Saturday']:<8} | {days['Sunday']:<6} |")
```

## Step 4: Flatten the Data into a Single List

```python
email_counts = []
for person_days in emails_data.values():
    email_counts.extend(person_days.values())
```

## Step 5: Calculate Average Rate (位)

```python
lambda_val = np.mean(email_counts)
print(f"\nCalculated Average Rate (位) across all persons and days: {lambda_val:.2f} emails/day")
```

## Step 6: Take k as User Input

```python
k_value = int(input("\nEnter the specific number of emails (k) to find the probability for: "))
```

## Step 7: Calculate Poisson Probability

```python
probability = poisson.pmf(k_value, lambda_val)
print(f"\nProbability of receiving exactly {k_value} emails in a day: {probability:.4f} ({probability*100:.2f}%)")
```

## Step 8: Plot the Poisson Distribution

```python
k_range = range(0, max(email_counts) + 10)
poisson_probs = [poisson.pmf(k, lambda_val) for k in k_range]
plt.figure(figsize=(10,6))
plt.bar(k_range, poisson_probs, color='orchid', alpha=0.8, edgecolor='black')
plt.xlabel("Number of Emails Received per Day (k)", fontsize=12)
plt.ylabel("Probability P(X=k)", fontsize=12)
plt.title(f"Poisson Distribution of Emails Received (位 = {lambda_val:.2f})", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
